## [Problem 1] Linear function

```python
import numpy as np
```

```python
x = np.arange(-50, 50.1, 0.1)
```

```python
y = 0.5*x+1
```

```python
x, y
```

## [Problem 2] Array combination

```python
xy = np.vstack((x, y)).T
xy.shape
```

## [Problem 3] Find the gradient

```python
x_diff = np.diff(x)
y_diff = np.diff(y)
```

```python
grad = y_diff/x_diff
```
